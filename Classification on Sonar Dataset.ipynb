```python
x=df.drop(columns="R")
y=df["R"]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression 
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(xtrain,ytrain)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
```

```python
from sklearn.svm import SVC
model=SVC (kernel="rbf",C=1)
model.fit (xtrain,ytrain)
```

```python
from sklearn.naive_bayes import GaussianNB
model= GaussianNB()
model.fit(x,y)
```

```python
PipelineDict = {0: 'Logistic Regression', 1: 'Random Forest', 2: 'Decision Tree', 3: 'SVC', 4: 'Gaussian'}
```

```python
mypipeline = [LogisticRegression(), RandomForestClassifier(), DecisionTreeClassifier(), SVC(C=1), GaussianNB()]
#fitting 
for mypipe in mypipeline:
    mypipe.fit(xtrain, ytrain)
```

```python
for i,model in enumerate(mypipeline):
    print("{} Test Accuracy: {}".format(PipelineDict[i],model.score(xtest,ytest)))
```
